# Automatic Natural Language Understanding

In <b><i>word sense disambiguation</i></b> we want to work out which sense of a word was intended in a given context.<br><br>
a. The thieves stole the paintings. They were subsequently sold.<br>
b. The thieves stole the paintings. They were subsequently caught.<br>
c. The thieves stole the paintings. They were subsequently found.<br>
Answering this question involves finding the <b><i>antecedent</i></b> of the pronoun they, either
thieves or paintings. Computational techniques for tackling this problem include <b><i>anaphora
resolution</i></b>—identifying what a pronoun or noun phrase refers to—and <b><i>semantic role labeling</i></b>—identifying how a noun phrase relates to the verb (as agent,
patient, instrument, and so on).</b>
